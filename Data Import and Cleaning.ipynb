```python
# now want to change those strange floats into integers - square footage, years, columns where there shouldn't be a decimal
# this is probably just a me OCD thing more than for actual modeling reasons, but think it just makes it all cleaner
kc_data.sqft_basement = kc_data.sqft_basement.astype('int64')
kc_data.yr_renovated = kc_data.yr_renovated.astype('int64')
```

```python
kc_data.info()
```

```python
# Looks good! Let's check some correlation

# import other libraries I will need
import seaborn as sns
import matplotlib.pyplot as plt
# magic!
%matplotlib inline 
```

```python
# Let's make a heat map!
# Set the style of the visualization
sns.set(style="white")

# Create a covariance matrix
corr = kc_data.corr()

# Generate a mask the size of our covariance matrix
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=0.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": 0.5})
```
